### Import libraries and extract data from API

```python
import requests
import urllib.request, json
from pandas.io.json import json_normalize
import pandas as pd
import geocoder

url = 'https://data.melbourne.vic.gov.au/resource/afyu-wy46.json'  

r_json = requests.get(url).json()
df = json_normalize(r_json)
df.head()
```

```python
df
```

### Only retain rows that have dogs offleash

```python
ind = df[df["status"] == "no"].index
df.drop(index=ind, inplace=True)
df.reset_index(drop = True, inplace=True)
df
```

### Extract co-ordinates details using Given multi-polygon co-ordinates
Extract the midpoint of latitude and longitude, using the closest and furthest co-ordinates

```python
longitude = []
latitude = []
df["Latitude"] = ["" for i in df.ref]
df["Longitude"] = ["" for i in df.ref]

for i in range(len(df)):
    lists = df["the_geom.coordinates"][i]
    for j in lists:
        for k in j:
            for l in k:
                latitude.append(l[1])
                longitude.append(l[0])
    print(min(latitude), max(latitude), min(longitude), max(longitude))
    df["Latitude"][i] = (min(latitude) + max(latitude))/2
    df["Longitude"][i] = (min(longitude) + max(longitude))/2
    latitude = []
    longitude = []
df        
```

### Extract Address, municipality and suburb using co-ordinates

```python
df["Address"] = ["" for i in df.ref]
df["Municipality"] = ["" for i in df.ref]
df["Suburb"] = ["" for i in df.ref]

for i in range(len(df)):
    g = geocoder.osm([df["Latitude"][i],df["Longitude"][i]], method='reverse')
    df["Address"][i] = g.json['address']
    df["Municipality"][i] = g.json["county"]
    df["Suburb"][i] = g.json["suburb"]
df
```

### Create categories column

```python
df["Categories"] = [[df["Municipality"][i], df["Suburb"][i], "Dog Walking Zones"] for i in range(len(df))]
df
```

### Write to file

```python
df.to_csv("../Data/Output files/DWZ.csv", index=False)
```
