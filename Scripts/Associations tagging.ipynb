### Import libraries and files

```python
import pandas as pd
df = pd.read_csv("../Data/Associations in melbourne.csv")
df
```

```python
df.columns
```

### Drop null values

```python
df.dropna(how='all', inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

### Add categories column

```python
df["Categories"] = [["Melbourne City Council", df["Suburb"][i], "Clubs"] for i in range(len(df))]
df
```

### Drop the unwanted column

```python
df.drop(columns=["Unnamed: 8"], inplace=True)
df
```

### Write to file

```python
df.to_csv("../Data/Output files/Clubs.csv", index=False)
```
